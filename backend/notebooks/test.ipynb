```python
import pandas as pd

# Basic dataset info
print("Rows:", len(df))
print("Columns:", len(df.columns))
print("Shape:", df.shape)

# Column names
print("\nColumns:")
print(df.columns.tolist())

# Event type distribution
print("\nEvent Type Counts:")
print(df["event_type"].value_counts(dropna=False))

# Event cause distribution
print("\nEvent Cause Counts:")
print(df["event_cause"].value_counts(dropna=False))

# Missing values
print("\nMissing Values:")
print(df.isnull().sum())

# Zone distribution
print("\nZone Counts:")
print(df["zone"].value_counts(dropna=False))

# Planned events analysis
planned = df[df["event_type"] == "planned"]

print("\nPlanned Events:", len(planned))
print("\nPlanned Event Causes:")
print(planned["event_cause"].value_counts(dropna=False))

# Convert datetime columns safely
df["start_datetime"] = pd.to_datetime(
    df["start_datetime"],
    errors="coerce"
)

df["resolved_datetime"] = pd.to_datetime(
    df["resolved_datetime"],
    errors="coerce"
)

# Resolution time in minutes
df["resolution_minutes"] = (
    df["resolved_datetime"] - df["start_datetime"]
).dt.total_seconds() / 60

print("\nResolution Time Statistics (minutes):")
print(df["resolution_minutes"].describe())

# Geographic coverage
print("\nUnique Junctions:", df["junction"].nunique())
print("Unique Zones:", df["zone"].nunique())
print("Unique Police Stations:", df["police_station"].nunique())

# Optional: Road closure statistics
print("\nRoad Closure Counts:")
print(df["requires_road_closure"].value_counts(dropna=False))

# Optional: Priority distribution
print("\nPriority Counts:")
print(df["priority"].value_counts(dropna=False))
```
